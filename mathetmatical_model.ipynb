### Problem description

The objective of the model is to provide a stowage plan for a set of cargoes with dynamic arrival times. Cargo is initially assumed to be uniform? 

Cargo can be placed in a slot on the deck and in order for the cargo to be unloaded there must exist a path from the slot to the exit ramp. It is assumed that cargo can be moved back and forth, but also sideways as long as there is ample space (how much can we assume?).

Arrival times are known, and are assumed to follow an exponential distribution. 

For cargo we will use estimates of revenue and loading time. If cargo is too delayed, it can be omitted from voyage at the loss of revenue.

Ideally, we want to minimize the time spent at the port, where fuel is consumed. In order for the objective function to have the same units, we will also have to estimate how much fuel is consumed per time. 

Additionally we also consider multiple ports, which means we want to minimuze the number of shifts required to unload cargo.

### Parameters, sets and variables needed

Inspired heavily by Alastair

To define problem:
- $C$ set of cargoes
- $P$ set of ports
- $I$ set of rows
- $J$ set of columns
- $I$ set of rows
- $N$ set of spots where cargo can be placed
- $E^I, E^J$ exit spots
- $C_p$ set of cargo to be unloaded at port $p\in P$
- $A$ arrival times of cargo

Parameters
- $C^{rev}$ revenue for transporting one piece of cargo
- $h_c$ time to handle one piece of cargo
- $C^{fuel}$ Fuel cost per time
- $C^{shift}$ pseudo-cost of shifting one cargo. 

Decisions
- $T = \sum_{c\in C}\sum_{(i,j)\in N} h_cx_{ijc} + \text{wait time}$: total time spent
- $\delta_{ijcp}$ 1 if cargo c in position (i,j) is shifted at port p

### Model notation

An object function that maximizes revenue:

$$
\min \quad TC^{fuel}\quad + \quad \sum_{p\in P}\sum_{c\in C}\sum_{(i,j)\in N} C^{shift}\delta_{ijcp}\quad -\quad \sum_{c\in C}\sum_{(i,j)\in N} C^{rev}x_{ijc} 

$$
